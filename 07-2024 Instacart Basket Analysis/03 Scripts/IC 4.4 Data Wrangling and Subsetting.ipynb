## 4. Find another identifier variable in the df_ords dataframe that doesn’t need to be included in your analysis as a numeric variable and change it to a suitable format.

Begin by reviewing data types for each variable. 
Order_id and user_id would not negatively impact analysis if changed to a non-numerical format. 
For this exercise I will change order_id to a string. 

```python
df_ords.dtypes #view data types for each variable
```

```python
#changes data type of 'order_id' in df_ords to a string and assigns it to the 'order_id' column in df_ords
df_ords['order_id'] = df_ords['order_id'].astype('str')
```

```python
#checks to ensure data type was successfully changed
df_ords.dtypes
```

The data type for order_id is now object (which is pandas way of representing a string).

## 5. Look for a variable in your df_ords dataframe with an unintuitive name and change its name without overwriting the dataframe.

Begin by viewing column names. Change 'order_dow' to something more intuitive like 'order_day_of_week'. Save this column name as a "new" column. 

```python
df_ords.columns #view column names
```

```python
#renames 'order_dow' to 'order_day_of_week' column in the df_ords dataframe. 
#this does not overwrite the dataframe because we did not reassign it or use the 'inplace=True' argument.
df_ords.rename(columns = {'order_dow' : 'order_day_of_week'})
```
